## 1. Probabilità

Verrà introdotto il concetto di probabilità con diversi approcci, focalizzando l'attenzione sulle applicazioni scientifiche. La teoria della probabilità trova applicazione in due ambiti principali:

* **fenomeni casuali** nella forma di esperimenti che possono essere replicati mantenendo le stesse condizioni, ma dai quali si potrà ottenere, ogni volta che l'esperimento viene ripetuto, un risultato diverso e non prevedibile;
* **ipotesi o fenomeni non noti o parzialmente noti**, la cui conoscenza potrà migliorare in base alle osservazioni di evidenze sperimentali. Quando una ipotesi acquisisce una credibilità molto grande rispetto alle altre ipotesi alternative, avremo una sua conferma, altrimenti una smentita. In questo senso, la teoria della probabilità è strettamente connessa con il **metodo scientifico** che prova teorie e principi sulla base di evidenze sperimentali.


## 2. Numeri casuali

Impareremo a generare con algoritmi sequenze di numeri (pseudo)casuali che seguono distribuzioni volute. 

Generare numeri casuali o pseudocasuali permette di realizzare esperimenti simulati complementari a quelle di laboratorio.

Per fare questo, introdurremo i **generatori di numeri pseudocasuali** e le tecniche che prendono in generale il nome di metodi **Monte Carlo**.

## 3. Approcci Bayesiano e frequentista alla probabilità

Avendo a disposizione generatori di numeri casuali, che ci forniscono un ambiente di laboratorio artificiale, partiremo dalla definzione di **probabilità classica** per poi introdurre due approcci complementari alla probabilità:
* l'approccio **Bayesiano**, o soggettivo
* l'approccio **frequentista**

Comprenderemo i campi di applicazione dei due metodi e i limiti di ciascuno.

Le proprietà generali della teoria della probabilità, qualsiasi sia la base concettuale sottostante, possono essere stabilite a partire dall'**approccio assiomatico dovuto a Kolmogorov**, valido per ogni approccio alla probabilità.

## 5. Stima dei parametri
In fisica sperimentale è necessario determinare i parametri ignoti di alcune distribuzioni in seguito all'osservazione di risultati sperimentali.
Questo problema prende il nome di **inferenza**, e si realizza in modo diverso negli approcci bayesiano e frequentista.

Comprenderemo cosa significa **misurare una grandezza fisica** applicando l'inferenza bayesiana o frequentista, e studieremo come assegnare correttamente un **errore** alla misura effettuata, e quindi come trasformare i dati sperimentali in conoscenza scientifica.

Gli intervalli di incertezza assegnate ad una misura hanno significato molto diverso negli approcci bayesiano e frequentista. Si parla, rispettivamente, di **intervalli di credibilità** oppure di **intervalli di confidenza**. Ne comprenderemo il significato ed impareremo come valutarli.

## 6. Test di ipotesi

Oltre alla misura di grandezze fisiche, i dati sperimentali consentono di discriminare tra più ipotesi quale siano vere e quali siano false, o, almeno, quali siano più plausibili, oppure descrivano meglio i dati, e quali meno.

Di nuovo, i due approcci bayesiano e frequentista forniscono metodi complementari per effettuare il test di ipotesi.

Diversi algoritmi numerici possono essere usati per discriminare diverse ipotesi, con prestazioni diverse. Alla fine del corso, daremo anche una breve introduzione agli algoritmi di **machine learning** che permettono di ottenere prestazioni vicine ai livelli ottimali previsti dalla teoria.

Oppure, in assenza di un segnale che evidenzi un nuovo venomeno che si vuole cercare, come si possa escluderne la presenza almeno per certi intervalli dei valori dei parametri presenti nei modelli ipotizzati. Ad esempio, si può escludere l'esistenza di una particella fino ad un certo valore della sua massa.
