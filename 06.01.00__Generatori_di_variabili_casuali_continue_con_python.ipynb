# Generatori di variabili casuali continue con python

La guida completa sui generatori disponibili si trova nella [documentazione](https://docs.python.org/3/library/random.html) della libreria ```random```.

Anche il pacchetto ```scipy.stats``` contiene diverse funzioni statistiche, descritte nella [documentazione](https://docs.scipy.org/doc/scipy/reference/stats.html).

````python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.stats import binom
from scipy.stats import poisson
import math
random.seed(1234)
````

````python
def plot_rnd(f, xmin, xmax, num = 100000):
    seq = [f() for i in range(num)]
    nbins = 100
    fig, ax = plt.subplots(figsize=(9, 6))
    n, bins, patches = ax.hist(seq, nbins, density=True, range=(xmin, xmax))
    plt.xlabel('x')
    plt.ylabel('freq.')
    plt.show()
````

````python
plot_rnd(lambda: random.uniform(-2, 2), -5, 5)
````

````python
plot_rnd(lambda: random.expovariate(2), 0, 4)
````

````python
plot_rnd(lambda: random.gauss(1, 0.5), -2, 4)
````

````python
plot_rnd(lambda: random.lognormvariate(1, 0.5),0, 20)
````

La distribuzione binomiale ed il corrispondente generatore random sono definiti nel modulo ```scipy.stats.binom```.
Vedere la [documentazione](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html).

````python
plt.figure(figsize=(9,6))
for p in np.linspace(0.1, 0.9, 9):
    n = 100
    seq = binom.rvs(n, p, size=10000)
    nbins = n+1
    plt.hist(seq, nbins, density=True, range=(-0.5, n + 0.5), alpha=0.5, label = "p = {:.2f}".format(p))
plt.xlabel("n", size=14)
plt.ylabel("frequency", size=14)
plt.title("Binomial distribution")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
````

La distribuzione di Poisson ed il suo generatore random sono definiti nel modulo ```scipy.stats.poisson```.
Vedere la [documentazione](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html).

````python
plt.figure(figsize=(9,6))
for nu in [1, 5, 10, 15]:
    seq = poisson.rvs(nu, size=10000)
    nbins = 31
    plt.hist(seq, nbins, density=True, range=(-0.5, nbins - 0.5), alpha=0.5, label = r"$\mu$ = {:.2f}".format(nu))
plt.xlabel("n", size=14)
plt.ylabel("frequency", size=14)
plt.title("Poisson distribution")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
````

## Teorema del limite centrale

Abbiamo $N$ variabili casuali indipendenti e distribuite secondo la stessa distribuzione, $x_i$, $i=1,\cdots,N$. <span style="color: blue">Se la loro varianza  **finita**</span>, la loro media $\bar{x}$ tender, per $N\rightarrow\infty$, ad avere una distribuzione normale.

La dimostrazione la omettiamo, ma proveremo a verificarla numericamente in alcuni casi.

Consideriamo che:

$\displaystyle\mathbb{V}\mathrm{ar}[\bar{x}] = \frac{\sigma^2}{N}$

Quindi, se vogliamo costruire una combinazione lineare degli $x_i$ che abbia la stessa deviazione standard $\sigma$ di ciascuna distribuzione individuale, dobbiamo definire:

$x^\star = \bar{x}\sqrt{N}$

In modo che:

$\displaystyle\mathbb{V}\mathrm{ar}[x^\star] = N\cdot\frac{\sigma^2}{N}=\sigma^2$

<span style="color: blue">Esempi di distribuzioni con varianza **infinita** si possono costruire partendo da funzioni che hanno "code" che decrescono come $1/x^2$ (es: $f(x) = 1/(1+x^2)$) per $x\rightarrow \pm\infty$. Funzioni del genere sono normalizzabili ($1/x^2$  integrabile), ma la varianza sar infinit, perch il valor medio di $x^2$, richiede l'integrazione di $x^2/x^2=1$, che non  integrabile.</span>


````python
from scipy.stats import norm
def plot_rnd_mean(f, n, sigma, xmin, xmax, num=10000):
    def xn():
        return np.array([f() for j in range(n)]).mean()*math.sqrt(n)
    seq = np.array([xn() for i in range(num)])
    nbins = 100
    fig, ax = plt.subplots()
    n, bins, patches = ax.hist(seq, nbins, density=True, range=(xmin, xmax))
    x=np.linspace(xmin, xmax, nbins)
    ax.plot(x, norm.pdf(x, scale=sigma))
    plt.show()
    print("std. dev x*: ", seq.std())
````

````python
from scipy.stats import uniform
u = uniform(loc=-1, scale=2)
sigma = u.std()
plot_rnd_mean(lambda: u.rvs(), 1, sigma, -3, 3)
print("std. dev x: ", sigma)
````

````python
plot_rnd_mean(lambda: u.rvs(), 2, sigma, -3, 3)
````

````python
plot_rnd_mean(lambda: u.rvs(), 3, sigma, -3, 3)
````

````python
plot_rnd_mean(lambda: u.rvs(), 10, sigma, -3, 3)
````

Definiamo una distribuzione custom.

````python
def rnd_custom():
    r = random.randint(0, 1)
    if r>0: return random.uniform(1, 2)
    else: return random.uniform(-2, -1)
````

Il valor medio sar sicuramente pari a zero, ma dobbiamo calcolare la deviazione standard di questa distribuzione. 

La PDF sar:

$\displaystyle f(x) = \left\{
\begin{array}{ll}
\frac{1}{2} & \mathrm{se} & -2<x<-1\,\,\mathrm{o}\,\,1<x<2\\
0 & \mathrm{altrove} &
\end{array}
\right.
$


La varianza  pari a:

$\displaystyle \mathbb{V}\mathrm{ar}[x] = \frac{1}{2}\left(
\int_{-2}^{-1}x^2\,\mathrm{d}x + \int_{1}^{2}x^2\,\mathrm{d}x
\right) =
\int_{1}^{2}x^2\,\mathrm{d}x =
\left[\frac{x^2}{2}
\right]^{2}_1 = \frac{4 - 1}{2} = 1.5
$


````python
sigma = 1.5
plot_rnd_mean(rnd_custom, 1, sigma, -5, 5)
````

````python
plot_rnd_mean(rnd_custom, 2, sigma, -5, 5)
````

````python
plot_rnd_mean(rnd_custom, 3, sigma, -5, 5)
````

````python
plot_rnd_mean(rnd_custom, 4, sigma, -5, 5)
````

````python
plot_rnd_mean(rnd_custom, 10, sigma, -5, 5)
````

````python
plot_rnd_mean(rnd_custom, 50, sigma, -5, 5)
````

Questa dimostrazione giustifica l'uso della distribuzione normale in molte applicazioni. Ogni volta che un effetto  la risultante della sovrapposizione di tanti effetti diversi, la distribuzione si potr approssimare con una gaussiana.

Questo  vero con tanti parametri che caratterizzano la risposta di un rivelatore, come la risoluzione nella misura di posizione, energia, ecc
